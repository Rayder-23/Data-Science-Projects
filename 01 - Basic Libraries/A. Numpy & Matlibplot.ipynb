```python
import numpy as np

# 1D Array
arr1 = np.array([1,2,3,4,5])
print("1D Array: ", arr1)

# 2D Array
arr2 = np.array([[1,2,3], [4,5,6]])
print("2D Array:\n", arr2)
```

```python
# Element-wise operations
# simple operations are easier, no need for looping to modify all elements

arr = np.array([10,20,30,40])

print("Add 5:", arr + 5)
print("Multiply by 2:", arr * 2)
print("Square:", arr ** 2)
```

```python
# Statistics and Random Numbers
data = np.random.randint(1, 100, size=10) # array of 10 random integers

print("Data:", data)
print("Mean:", np.mean(data))
print("Standard Deviation:", np.std(data))
print("Maximum:", np.max(data))
```

```python
# Matrix Operations
A = np.array([[1,2], [3,4]])
B = np.array([[5,6], [7,8]])

print("Matrix Mutliplication:\n", np.dot(A,B))

print("Tranpose of A:\n", A.T)
```

```python
# Numpy Integration with Visualisation

# Basic Line Plot
import matplotlib.pyplot as plt

x = np.linspace(0, 10 ,100) # 100 points btwn 1 and 100 (evenly spaced)
y = np.sin(x)

plt.plot(x, y)  # Line plot
plt.title("Sine Wave")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.show() # A smooth sine wave
```

```python
# Scatter Plot

x = np.random.rand(50) # random x values
y = np.random.rand(50) # random y values

plt.scatter(x, y, color="red", marker="o")
plt.title("Random Scatter Plot")
plt.show()
```

```python
# Bar Chart

categories = ["A","B","C","D"]
values = [10, 24, 36, 18]

plt.bar(categories, values, color="green")
plt.title("Bar Chart Example")
plt.show()
```

```python
# Multiple Plots in One Figure

x = np.linspace(0, 10, 100)

plt.figure(figsize=(10, 5)) # bigger figure

plt.subplot(1, 2 , 1) # 1 row, 2 columns, first subplot
plt.plot(x, np.sin(x), label="Sine")
plt.legend()

plt.subplot(1, 2, 2) # second subplot
plt.plot(x, np.cos(x), color="orange", label="Consine")
plt.legend()

plt.show()
```

```python
# Histogram

data = np.random.randn(1000) # 1000 random values from normal distribution

plt.hist(data, bins=30, color="blue", alpha=0.7)
plt.title("Histogram Example")
plt.show()
```
